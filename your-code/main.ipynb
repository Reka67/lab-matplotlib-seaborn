#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x, y, color='green',linestyle='dashed',linewidth=5, markersize=5)
axes[1].plot(x, z, 'o', linewidth=2, markersize=5)
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
fig, axes = plt.subplots(1,2,figsize=(10,10))

axes[0].plot(x, y, color='green',label="y = x*2",linestyle='dashed',linewidth=5, markersize=5)
axes[1].plot(x, z, 'v', label="y = x**2" ,linewidth=2, markersize=5)
axes[0].set_title("linear", fontsize=20)
axes[1].set_title("exponential", fontsize=15)

for ax in axes:
    ax.legend()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
activities=["Minutes Lightly Active","Minutes Fairly Active", "Minutes Very Active" ]
colors=["yellowgreen","olivedrab","darkolivegreen"]
fig8, axes8 = plt.subplots(1,3, figsize=(15,5))
for i in range(0,3):
    plt.sca(axes8[i]) 
    plt.scatter(data = fitbit, x = activities[i], y="Steps", c=colors[i])
    axes8[i].set_title(activities[i], fontsize=18)
```
